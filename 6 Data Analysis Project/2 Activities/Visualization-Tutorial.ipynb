## Line Plot


We will start with a basic line chart, which includes an x-axis that identifies the objects we want to compare and a y-axis that includes the values for each object.

In the following example, we use Matplotlib's `plot` function to map the data items to the chart appropriately. 

We first define the axes as lists of values. Here, we use hard-coded values, but we could also use lists automatically generated from a dataset. Note that each list must be the same length. Matplotlib will map each item in each list to the corresponding item in the other list to generate the line chart. If the lists are of different lengths, Python will throw an error when generating the chart.

We then use the `plot` function to generate the chart. This function plots each data point based on the values in the x and y axes, and then connects the data points with a line.

This function also supports a wide range of customization, allowing us to specify format features such as line color, marker type, and what style line the line chart will use.

For more information about this function, see Matplotlib's page, [matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

```python
import matplotlib.pyplot as plot # Import the plot module from matplotlib
import numpy as np #Import numpy to manipulate arrrays

x = ["Ford", "Mazda", "Toyota", "Dodge", "Chrysler"] # create a list with the x values 
y = [0, 3, 2, 5, 4] # create a list with the y values 

# use the plot function within the plot module to plot a simple line plot
plot.plot(x, y,color='green', marker='o', linestyle='dashed') 
    # color: color of the line plot
    # marker: the marker used to locate the (x,y) values
    # linestyle: the line type for the line plot
    
plot.show() # display the line plot 
```

## Histogram Plot
A histogram is a type of column chart that shows the distribution of a collection of values, grouped by categories or **bins**. 

In a histogram, the x-axis represents the bins, while the y-axis represents the number of items in each bin. In the code, we provide a list of values. Using the default settings, the `hist` function automatically groups distinct values in the list and generates a plot that shows how many times each distinct value appears in the list.

The following example shows a basic histogram using minimal settings. We have a list of numbers, where each number is in the range 0 to 4. The `hist` function groups the numbers and counts the number of occurrences for each distinct value. The resulting graph shows each of the original numbers (0-4) on the x-axis, while the y-axis represents the number of times each of those numbers appears in the list.


```python
import matplotlib.pyplot as plot
import numpy as np

list1 = [4, 3, 3, 3, 1, 3, 0, 4, 0, 4, 4, 0, 3, 1, 1] 

plot.hist(list1)

plot.show()
```

The following example includes two lists of numbers. Each list includes 15 values, where each value is an integer between 0 and 4. The distribution of the values is different for each list, however.

We use the `subplots` function to generate a 1x3 grid that can display three charts on the same row. We then generate a separate chart for each list, along with a third chart that includes both lists in a single chart.

In the resulting charts, note that values on the x-axis correspond to the values in the lists, while the values on the y-axis correspond to the number of times each value appears in the list.
* In `list1`, the number 3 appears five times, but the number 2 does not appear at all.
* In `list2`, the number 0 appears five times, but the number 1 only appears once.

```python
import matplotlib.pyplot as plot
import numpy as np

list1 = [4, 3, 3, 3, 1, 3, 0, 4, 0, 4, 4, 0, 3, 1, 1] 
list2 = [3, 0, 1, 2, 2, 0, 0, 0, 0, 3, 3, 2, 3, 4, 1]

# we use the subplots function within the plot module to create multiple visualizations within the same figure 
fig, axs = plot.subplots(1, 3, figsize=(15, 5)) 
# we use a grid of 1x3 graphs (which means three graphs organized horizontally)
# we use figsize to customize the size of each graph 

axs[0].hist(list1)  #using axs[0] we can access the first figure and use hist to plot the histogram
axs[1].hist([list2]) #using axs[1] we can access the second figure and use hist to plot the histogram.
axs[2].hist([list1, list2]) #axs[2] uses the third figure to display both lists in one chart

plot.show()
```

## Scatter Plot 
A scatter plot maps individual values as dots on a grid. Clumps of dots allow the user to identify trends in the data, while isolated dots can help identify outliers from those trends.

The `scatter` function requires two lists of data to compare. In this example, we use two lists of random numbers, but we could instead start with lists of the values we want to chart.

The NumPy `random.seed` function generates a predictable random number, which is then used to generate the same set of numbers each time the script runs.

We create a NumPy matrix of 2x100, which will generate two lists of 100 numbers each, and we print out those lists.

We then use the `scatter` function to create the scatter points. The result shows the distribution of the random numbers in a range based on the values in the lists. 

```python
import matplotlib.pyplot as plot
import numpy as np

np.random.seed(1000) #create a custom seed for generating random numbers 
data = np.random.randn(2, 100) #create a numpy matrix of 2x100 of random numbers
print(data)

plot.scatter(data[0], data[1]) #create the scatter plot using the first 100 values as x and second 100 values as y 
plot.show()
```

## Histogram 2-D Plot

A histogram 2-D plot is essentially a merger of a histogram and a scatter plot. It groups similar values in two arrays and displays the results in a grid, using the same logic used in a scatter plot.

In the example below, we generate two arrays of 100 random numbers, starting with a random seed. We then use `hist2d` to plot the values into a 2-D histogram.

```python
import matplotlib.pyplot as plot
import numpy as np

np.random.seed(19680801)
data = np.random.randn(2, 100)

plot.hist2d(data[0], data[1]) 

plot.show()
```

## Pie Chart 
A pie chart shows how a single larger value is split across smaller sections.

In the following example, we create a pie chart using two lists: the `labels` list identifies the sections of the chart, and the `sizes` list provides the value for each section as a percentage of the whole. As with a line plot, both lists should have the same number of values, and the `pie` function maps each label to its corresponding value.

We also include an exploding section, which appears to be pulled out of the rest of the pie. We use the `explode` feature to identify which piece should be pulled out (in this case, the second one listed in each of the lists), and the value determines how far it explodes. 

We then use parameters of the `pie` function to format the chart, using the values from `labels` as the labels on the chart, and formatting the values as percentages. The `shadow` option creates a 3-D effect, and `startangle` determines how the pieces of the chart will be laid out. 

Finally, we specify that the axes of the chart should be equal, so that the pie is a circle.

> Note: In the example below, the values in the `sizes` list happen to add up to 100, so each value represents that percentage of the whole. If these values do not add up to 100, `pie` will automatically distribute them based on their percentage of the actual total.

```python
import matplotlib.pyplot as plot

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Electronics', 'Clothing', 'Home Goods', 'Cleaning Supplies']
sizes = [15, 30, 45, 10]
explode = (0, .25, 0, 0)  # only "explode" the 2nd slice (i.e. 'Clothing')

plot.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plot.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plot.show()
```

## Combine Several Visualizations
When analyzing large datasets, it can be useful to have several different charts to present different aspects of the data in a single view.
	
The example below shows how we can generate multiple visualizations based on the same two datasets. Here, we generate two datasets, each with 100 random values, and then use those datasets to generate a histogram, a scatter plot, a line chart, and a 2-D histogram.

The `subplots` function allows us to create a grid that will display all four charts in the same page, and we assign each chart to a specific area in that grid.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```
